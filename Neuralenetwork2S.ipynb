**Hebian learning model implementing**

```python
import numpy as np
q=np.array([1,-1,0,0.5])
print('Intial weights', q)
w=np.transpose(q)
c=1
x1 = np.array([1, -2,1.5, 0])
x2 = np.array([1, -0.5, -2, -1.5])
x3 = np.array([0, 1, -1, 1.5])
def signum(x):
    return np.sign(x)
net1 = np.dot(w,x1)
o1 = signum(net1)
dw = c*o1*x1
w=w+dw
net2 = np.dot(w,x2)
o2=signum(net2)
dw = c*o2*x2
w=w+dw
net3 = np.dot(w,x3)
o3=signum(net3)
dw = c*o3*x3
w=w+dw
print('Ater iterating 4 times updated weights are', w)

```
